```python
"""
Grouping dataframe based on the Pclass to fil the missing values in Age column,
because here the more number of passenger travelling in class 3, if we apply global median it may arase the hidden patterns, 
right approach is to apply group wise median
"""

df.groupby(['Pclass','Sex'])['Age'].median() # grouped based on two classes to fill missing age groups
```

```python
df['Age'] = df.groupby(['Pclass', 'Sex'])['Age'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
# Created a new column deck which helps to extract the deck platforms which is directly related to PClass and Survived
df['Deck'] = df['Cabin'].str.get(0).fillna('Unknown')
df = df.drop(columns=['Cabin'])
```

```python
# Insights

"""
1.Females had significantly higher survival rates than males.
2.First-class passengers had much higher survival rates than third-class passengers.
3.Deck information provided additional survival-related insights beyond passenger class.
4.Embarked was correlated with Pclass but still showed some independent relationship with survival.
5.Small families (1–2 members) had higher survival rates than passengers traveling alone or in very large families.
6.Fare was strongly correlated with Pclass.
7.Within first class, passengers paying higher fares tended to survive more often.
8.Fare added little information within third class.
9.Most passengers paid relatively low fares, producing a heavily right-skewed distribution.
10.Socio-economic status (Pclass + Fare) was a major factor influencing survival.
"""
```
